```python
import pandas as pd
```

```python
df = pd.read_csv("Resources/EPL_20_21.csv")
df.head(20)
```

```python
df_clean = df[["Name","Goals"]].sort_values(by='Goals', ascending=False)
df_top10goals=df_clean.head(10)
df_top10goals
```

```python
df_top10goals.to_csv(r'top10goals.csv')
```

```python
df_clean1 = df[["Name","Assists"]].sort_values(by='Assists', ascending=False)
df_top10assists=df_clean1.head(10)
df_top10assists
```

```python
df_top10assists.to_csv(r'top10assists.csv')
```

```python
df_clean2 = df[["Name","Passes_Attempted"]].sort_values(by='Passes_Attempted', ascending=False)
df_top10passes=df_clean2.head(10)
df_top10passes
```

```python
df_top10passes.to_csv(r'top10passes.csv')
```
